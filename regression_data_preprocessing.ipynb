#### 1-2. Height 이상치 및 결측치 처리

```python
# 이상치 -> 0.5 이상
# length나 diameter가 동일한 값을 가진 데이터들의 평균값을 활용
df[df['Height'] > 0.5]
```

```python
df.loc[1417, 'Height'] = df[df['Diameter'] == 0.455]['Height'].mean().round(4)
df.loc[1417, 'Height']
```

```python
df.loc[2051, 'Height'] = df[df['Diameter'] == 0.355]['Height'].mean().round(4)
df.loc[2051, 'Height']
```

```python
# 결측치 처리 -> 값이 0
# # length나 diameter가 동일한 값을 가진 데이터들의 평균값을 활용
df[df['Height'] == 0]
```

```python
df.loc[1257, 'Height'] = df[df['Diameter'] == 0.34]['Height'].mean().round(4)
df.loc[3996, 'Height'] = df[df['Diameter'] == 0.23]['Height'].mean().round(4)
```

```python
df[df['Height'] == 0]
```
